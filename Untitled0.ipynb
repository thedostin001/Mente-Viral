```python
from transformers import pipeline
import random

# Lista de temas
temas = [
    "Finanzas personales",
    "Tecnología",
    "Desarrollo personal",
    "Curiosidades científicas",
    "Espiritualidad y misterio"
]

# Selecciona un tema aleatorio
tema = random.choice(temas)

# Configura el generador de texto
generador = pipeline("text-generation", model="gpt2")

# Genera el artículo
entrada = f"Artículo sobre {tema}:"
resultado = generador(entrada, max_length=500, num_return_sequences=1)

# Muestra el artículo generado
print(f"Tema seleccionado: {tema}\n")
print(resultado[0]['generated_text'])
```

```python
# Guarda el artículo en un archivo
with open("articulo_generado.txt", "w") as archivo:
    archivo.write(resultado[0]['generated_text'])

print("El artículo ha sido guardado como 'articulo_generado.txt'")
```
